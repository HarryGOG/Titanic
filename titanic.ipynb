# E) Feature Engineering

```python
import pandas as pd
import numpy as np

#we load  data
train = pd.read_csv('train.csv', sep=',')
test = pd.read_csv('test.csv', sep=',')

#WE combine the two dataset
test['Survived'] = float('nan')

data = pd.concat([train,test])
data
```

```python
import re

#spliting name by using ','  or '.' characters
data['Title'] = data['Name'].map(lambda x:  re.split(r',|\.', x)[1].strip())

#Statistics in the fiels Title
data.Title.value_counts()
```

```python
#We can combined some values in the Title field 
data['Title'][data['Title']=='Mme'] ='Mlle'    
data['Title'] = data['Title'].map(lambda x: 'Sir' if x in ['Capt', 'Don', 'Major']  else x  )
data['Title'] = data['Title'].map(lambda x: 'Lady' if x in ['Dona', 'Lady', 'the Countess', 'Jonkheer']  else x  )

#Statistics in the fiels Title
data.Title.value_counts()
```

```python
#SibSb and Parch indicate the number of family members the passenger is travelling with
#letâ€™s combine the two variables into a new one, FamilySize:

data['FamilySize'] = data['SibSp'] + data['Parch'] + 1

#Combining the Surname with the family size to create FamilyID 
#spliting name by using ','  or '.' characters
data['Surname'] = data['Name'].map(lambda x:  re.split(r',|\.', x)[0].strip())
data['FamilyID'] = data['FamilySize'].apply(str) + data['Surname']

famIDs = data['FamilyID'].value_counts().to_dict()
data['FamilyID'] = [ 'Small' if famIDs.get(x)<=3  else x for x in data['FamilyID']]
data
```

```python
#Descriptive statistics
print data.describe(include = [np.number]), "\n" 
print data.describe(include = ['O']), "\n"

#We count the number of NA values in each column
print "Number of NA/null values: \n", data.isnull().sum()
```

```python
#we have 2 missing values in the fields Embarked. We fill it with the value 'S'
#data.Embarked.unique()
data['Embarked'][data['Embarked'].isnull()] = 'S'
data['Embarked'].isnull().sum()
```

```python
#we have 1 missing value in the fields Fare. We fill it with the median

Median_fare = data['Fare'].median()
data['Fare'][data['Fare'].isnull()] = Median_fare  

#Control
data['Fare'].isnull().sum()
```

```python
# we transform string data to float, because some algorithm doesn't use string data
data['Sex2'] = [0 if x =='male' else 1 for x in data['Sex']]

#data.Embarked.unique()
data['Embarked2'] = float('nan')

mydict = {'S': 0, 'C': 1, 'Q': 2}
#mydict['this']
data['Embarked2'] = data.Embarked.map(lambda x: mydict[x])
data.Embarked2.unique()
```

```python
from sklearn import preprocessing

le_FamilyID = preprocessing.LabelEncoder()
#to convert into numbers familyid 

data['FamilyID2'] = le_FamilyID.fit_transform(data.FamilyID)
#to convert back


le_Title = preprocessing.LabelEncoder()
#to convert into numbers Title 

data['Title2'] = le_Title.fit_transform(data.Title)
#to convert back
data
```

```python
#In the age field, 263 values out of 1309 were missing. we fill it by the median value

#Median_age = data['Age'].median()
#data['Age'][data['Age'].isnull()] = Median_age  

#Control
#data['Age'].isnull().sum()
```

```python
#In the age field, 263 values out of 1309 were missing. we fill it by predicting the value of age
from sklearn.tree import DecisionTreeClassifier, export_graphviz

dat = data[['Pclass', 'Sex2',  'Age', 'SibSp', 'Parch', 'Fare', 'Embarked2', 'Title2', 'FamilySize', 'FamilyID2']].dropna()
dat2 = dat[['Pclass', 'Sex2', 'SibSp', 'Parch', 'Fare', 'Embarked2', 'Title2', 'FamilySize', 'FamilyID2']]

dt = DecisionTreeClassifier(random_state=99)
dt.fit(dat2, np.asarray(dat['Age'], dtype="|S6") )


#dat3 = data[['Pclass', 'Sex2', 'SibSp', 'Parch', 'Fare', 'Embarked2', 'Title2', 'FamilySize', 'FamilyID2']][data['Age'].isnull()]
data['Age'][data['Age'].isnull()] = pd.to_numeric(dt.predict(data[['Pclass', 'Sex2', 'SibSp', 'Parch', 'Fare', 'Embarked2', 'Title2', 'FamilySize', 'FamilyID2']][data['Age'].isnull()]))

data['Age'].isnull().sum()
```

```python
#We split data into train and test data
train =  data[data.Survived.isnull() == False]
print train.count(), '\n'

test = data[data.Survived.isnull()]
print test.count()
```
